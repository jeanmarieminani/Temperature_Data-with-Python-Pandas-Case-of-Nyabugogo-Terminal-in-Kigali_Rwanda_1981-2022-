```python
# Plotting 'YEAR' column vs. columns'JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC' 
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(data['YEAR'], data['JAN'], color='black', label='January')
ax.scatter(data['YEAR'], data['FEB'], color='blue', label='February')
ax.scatter(data['YEAR'], data['MAR'], color='green', label='March')
ax.scatter(data['YEAR'], data['APR'], color='red', label='April')
ax.scatter(data['YEAR'], data['MAY'], color='purple', label='May')
ax.scatter(data['YEAR'], data['JUN'], color='orange', label='June')
ax.scatter(data['YEAR'], data['JUL'], color='brown', label='July')
ax.scatter(data['YEAR'], data['AUG'], color='pink', label='August')
ax.scatter(data['YEAR'], data['SEP'], color='gray', label='September')
ax.scatter(data['YEAR'], data['OCT'], color='cyan', label='October')
ax.scatter(data['YEAR'], data['NOV'], color='magenta', label='November')
ax.scatter(data['YEAR'], data['DEC'], color='yellow', label='December')
ax.set_xticks(range(1980, 2024, 4))
ax.set_xlabel('Year')
ax.set_title('Nyabugogo_Kigali Monthly Average Temperature 1981-2022')
ax.set_ylabel('Monthly Average Temperature')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
